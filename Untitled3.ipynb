```python
import cv2

def process_image(image_path):
    # Step 1: Load the image
    image = cv2.imread(image_path)
    image_resized = cv2.resize(image, (256, 256))  # Resize to match DehazeNet input size

    # Step 2: Defogging using DehazeNet
    defogged_image = dehazenet.predict(image_resized[None, ...] / 255.0)[0]  # Normalize input

    # Convert defogged image to uint8 for YOLOv8
    defogged_image = (defogged_image * 255).astype("uint8")

    # Step 3: Object Detection using YOLOv8
    results = yolov8.predict(source=defogged_image, save=False)

    return defogged_image, results
```

```python
# Path to a sample foggy image
# image_path = r"F:\D-drive Storage\MATLAB\bin\MatlabDefogging\maritime-defog\Test_Data\V_07_01_0014.jpg"
# image_path = r"C:\Users\atul toppo\Documents\7thSem_Minor_Project\Dataset\OTS\part4\7238_1_0.16.jpg"
image_path = r'C:\Users\atul toppo\Documents\7thSem_Minor_Project\Dataset\OTS\part4\7238_1_0.16.jpg'

# Process the image
defogged_image, results = process_image(image_path)

import matplotlib.pyplot as plt

# Display defogged image using Matplotlib
plt.imshow(cv2.cvtColor(defogged_image, cv2.COLOR_BGR2RGB))
plt.title("Defogged Image")
plt.axis("off")
plt.show()

# Display YOLOv8 detection results
results[0].show()  # Shows detection on the defogged image
```
