```python
# Dataset paths
ots_hazy_path = r'C:\Users\atul toppo\Documents\7thSem_Minor_Project\Dataset\OTS\part4'
ots_clear_path = r'C:\Users\atul toppo\Documents\7thSem_Minor_Project\Dataset\OTS\clear\clear'

# Load image paths
ots_hazy_imgs = [os.path.join(ots_hazy_path, f) for f in os.listdir(ots_hazy_path) if f.endswith('.jpg') or f.endswith('.png')]
ots_clear_imgs = [os.path.join(ots_clear_path, f) for f in os.listdir(ots_clear_path) if f.endswith('.jpg') or f.endswith('.png')]

# Split for validation (considering large size, using smaller validation set)
train_hazy, val_hazy = train_test_split(ots_hazy_imgs, test_size=0.2, random_state=42)
train_clear, val_clear = train_test_split(ots_clear_imgs, test_size=0.2, random_state=42)
```

```python
class DehazeGenerator(keras.utils.Sequence):
    def __init__(self, hazy_imgs, clear_imgs, batch_size, dim=(256, 256), n_channels=3, shuffle=True, mode='random_sample'):
        self.hazy_imgs = hazy_imgs
        self.clear_imgs = clear_imgs
        self.batch_size = batch_size
        self.dim = dim
        self.n_channels = n_channels
        self.shuffle = shuffle
        self.mode = mode  # 'random_sample' or 'fixed_mapping' (not implemented for fixed mapping)
        self.on_epoch_end()

    def __len__(self):
        return int(np.ceil(len(self.hazy_imgs) / float(self.batch_size)))

    def on_epoch_end(self):
        self.indexes = np.arange(len(self.hazy_imgs))
        if self.shuffle == True:
            np.random.shuffle(self.indexes)

    def __getitem__(self, idx):
        indexes = self.indexes[idx*self.batch_size:(idx+1)*self.batch_size]
        
        hazy_batch = [self.hazy_imgs[k] for k in indexes]
        
        # **Handle Clear Images Based on Selected Mode**
        if self.mode == 'random_sample':
            clear_batch = np.random.choice(self.clear_imgs, size=len(hazy_batch), replace=True)
        elif self.mode == 'fixed_mapping':  # **Not Implemented for Fixed Mapping**
            # TODO: Implement fixed mapping logic if required
            pass
        else:
            raise ValueError("Invalid mode. Please choose 'random_sample' or 'fixed_mapping' (although 'fixed_mapping' is not implemented).")
        
        # **Generate Data**
        X, y = self.__data_generation(hazy_batch, clear_batch)
        
        return X, y

    def __data_generation(self, hazy_batch, clear_batch):
      X = np.zeros((self.batch_size, *self.dim, 3))  # Ensure 3 channels (RGB)
      y = np.zeros((self.batch_size, *self.dim, 3))  # Ensure 3 channels (RGB)
    
      for i, (haz, cl) in enumerate(zip(hazy_batch, clear_batch)):
        # **Load and Resize Images**
        haz_img = Image.open(haz)
        cl_img = Image.open(cl)
        haz_img = haz_img.resize(self.dim)
        cl_img = cl_img.resize(self.dim)
        
        # **Remove Alpha Channel (if present) and Convert to Numpy Arrays**
        if haz_img.mode == 'RGBA':
            haz_img = haz_img.convert('RGB')  # Remove alpha channel
        if cl_img.mode == 'RGBA':
            cl_img = cl_img.convert('RGB')  # Remove alpha channel
        
        haz_img = np.array(haz_img) / 255.0
        cl_img = np.array(cl_img) / 255.0
        
        # **Store the Resized and Normalized Images**
        X[i,] = haz_img
        y[i,] = cl_img
        
      return X, y

# **Initialize Custom Generator with 'random_sample' Mode**
batch_size = 4
train_gen = DehazeGenerator(train_hazy, train_clear, batch_size=batch_size, mode='random_sample')
val_gen = DehazeGenerator(val_hazy, val_clear, batch_size=batch_size, shuffle=False, mode='random_sample')
```

```python
def DehazeNet(input_shape=(256, 256, 3)):
    inputs = keras.layers.Input(shape=input_shape)
    
    # Encoder
    conv1 = keras.layers.Conv2D(3, (1, 1), activation='relu', padding='same')(inputs)
    conv2 = keras.layers.Conv2D(3, (3, 3), activation='relu', padding='same')(conv1)
    conv3 = keras.layers.Conv2D(3, (5, 5), activation='relu', padding='same')(conv2)
    conv4 = keras.layers.Conv2D(3, (3, 3), activation='relu', padding='same')(conv3)
    conv5 = keras.layers.Conv2D(3, (1, 1), activation='tanh', padding='same')(conv4)
    
    # Skip Connection
    outputs = keras.layers.Add()([conv5, inputs])
    
    model = keras.Model(inputs=[inputs], outputs=[outputs])
    return model

# Initialize DehazeNet model
dehaze_model = DehazeNet()
dehaze_model.summary()
```

```python
# Compile the model
dehaze_model.compile(optimizer='adam', loss='mean_squared_error', metrics=['mean_absolute_error'])

# Define callbacks for training
from keras.callbacks import ModelCheckpoint, EarlyStopping, ReduceLROnPlateau
checkpoint = ModelCheckpoint("dehaze_model.h5", monitor='val_loss', verbose=1, save_best_only=True, mode='min')
early_stop = EarlyStopping(monitor='val_loss', min_delta=0.001, patience=5, mode='min', verbose=1)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=3, min_lr=0.001, verbose=1)

# Lists to track training and validation metrics for visualization
train_loss = []
val_loss = []
train_mae = []
val_mae = []
psnr_values = []
ssim_values = []
```

```python
# Train the model
history = dehaze_model.fit(train_gen, 
                           validation_data=val_gen, 
                           epochs=10, 
                           callbacks=[checkpoint, early_stop, reduce_lr])

# Update lists with training and validation metrics
train_loss.extend(history.history['loss'])
val_loss.extend(history.history['val_loss'])
train_mae.extend(history.history['mean_absolute_error'])
val_mae.extend(history.history['val_mean_absolute_error'])

# Calculate PSNR and SSIM for validation set (example for the last batch)
for haz, clear in zip(val_gen, val_gen):
    predicted = dehaze_model.predict(haz)
    for p, c in zip(predicted, clear):
        psnr_values.append(compare_psnr(c, p))
        ssim_values.append(compare_ssim(c, p, multichannel=True))
    break  # Calculate for one batch for demonstration
```

```python
import tensorflow as tf
from tensorflow import keras

def ImprovedDehazeNet(input_shape=(256, 256, 3)):
    inputs = keras.layers.Input(shape=input_shape)
    
    # Multi-scale feature extraction
    branch1 = keras.layers.Conv2D(32, (1, 1), activation='relu', padding='same')(inputs)
    branch2 = keras.layers.Conv2D(32, (3, 3), activation='relu', padding='same')(inputs)
    branch3 = keras.layers.Conv2D(32, (5, 5), activation='relu', padding='same')(inputs)
    multi_scale = keras.layers.Concatenate()([branch1, branch2, branch3])
    
    # Feature refinement
    conv1 = keras.layers.Conv2D(64, (3, 3), activation='relu', padding='same')(multi_scale)
    conv1 = keras.layers.BatchNormalization()(conv1)
    conv2 = keras.layers.Conv2D(64, (3, 3), activation='relu', padding='same')(conv1)
    conv2 = keras.layers.BatchNormalization()(conv2)
    
    # Output layer
    outputs = keras.layers.Conv2D(3, (1, 1), activation='sigmoid', padding='same')(conv2)
    
    # Skip connection
    outputs = keras.layers.Add()([outputs, inputs])
    
    model = keras.Model(inputs=[inputs], outputs=[outputs])
    return model

# Initialize and summarize the improved model
dehaze_model = ImprovedDehazeNet()
dehaze_model.summary()
```
